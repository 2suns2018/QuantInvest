```python
import akshare as ak
import pandas as pd
import numpy as np
import calendar
%matplotlib inline

#设置完整显示数据
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 
              500)
pd.set_option('display.width', 1000)
pd.set_option('display.float_format', '{:,.2f}'.format)
```

```python
stock_register_kcb_df = ak.stock_register_kcb()
stock_register_cyb_df = ak.stock_register_cyb()
stock_register_kcb_df = stock_register_kcb_df.append(stock_register_cyb_df, ignore_index=True)
```

```python
print(stock_register_kcb_df)
```

```python
stats = pd.to_datetime(stock_register_kcb_df["受理日期"]).dt.strftime("%b").value_counts()
month_names = calendar.month_abbr[1:]
stats_by_month = stats.reindex(month_names, fill_value=0)
stats_by_month
```

```python
(stats_by_month/stats_by_month.sum())*100
```

```python
stats_brokerage = stock_register_kcb_df["保荐机构"].value_counts().rename_axis('保荐机构').to_frame('受理申报数')
stats_brokerage
```

```python
(stats_brokerage['受理申报数']/stats_brokerage['受理申报数'].sum())*100
```

```python
stock_registered = stock_register_kcb_df[stock_register_kcb_df['审核状态'] == '注册生效']
stock_registered['拟上市地点'].value_counts()
```

```python
stock_registered = stock_registered["保荐机构"].value_counts().rename_axis('保荐机构').to_frame('注册生效数')
stock_registered
```

```python
applied_to_registered = pd.concat([stats_brokerage,stock_registered], axis=1)
applied_to_registered['通过率'] = applied_to_registered['注册生效数'].div(applied_to_registered['受理申报数']) * 100
applied_to_registered
```
